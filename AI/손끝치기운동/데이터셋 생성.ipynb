```python
import cv2
import mediapipe as mp
import numpy as np
import csv
import os

# MediaPipe 초기화
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
mp_drawing_styles = mp.solutions.drawing_styles

# CSV 파일 경로
wrong_file = 'wrong_data.csv'

# CSV 파일이 없으면 헤더 작성
if not os.path.exists(wrong_file):
    with open(wrong_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        # 손 랜드마크는 21개의 점으로 구성되어 있으므로 21*3 = 63개의 값이 있음
        header = [f'x{i}' for i in range(21)] + [f'y{i}' for i in range(21)] + [f'z{i}' for i in range(21)]
        writer.writerow(header)

# 웹캠을 통해 영상을 캡처합니다.
cap = cv2.VideoCapture(0)
with mp_hands.Hands(
    model_complexity=0,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
  
  while cap.isOpened():
    success, image = cap.read()
    if not success:
      print("카메라를 찾을 수 없습니다.")
      continue

    # 성능 향상을 위해 이미지 작성 가능성을 비활성화합니다.
    image.flags.writeable = False
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(image)

    # 이미지 작성 가능성을 다시 활성화합니다.
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    
    if results.multi_hand_landmarks:
      for hand_landmarks in results.multi_hand_landmarks:
        mp_drawing.draw_landmarks(
            image,
            hand_landmarks,
            mp_hands.HAND_CONNECTIONS,
            mp_drawing_styles.get_default_hand_landmarks_style(),
            mp_drawing_styles.get_default_hand_connections_style())

        # 랜드마크 데이터 추출
        landmarks = [[lm.x, lm.y, lm.z] for lm in hand_landmarks.landmark]
        landmarks_flat = np.array(landmarks).flatten().tolist()

        # CSV 파일에 데이터 추가
        with open(wrong_file, mode='a', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(landmarks_flat)
        print("Wrong 데이터 저장")

    # 보기 편하게 이미지를 좌우 반전합니다.
    cv2.imshow('MediaPipe Hands', cv2.flip(image, 1))
    
    # ESC 키를 누르면 종료합니다.
    if cv2.waitKey(5) & 0xFF == 27:
      break

cap.release()
cv2.destroyAllWindows()
```
